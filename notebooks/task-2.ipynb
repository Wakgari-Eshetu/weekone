```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pynance as pn
import talib
```

```python
#Load & Prepare Stock Data
df = pd.read_csv('../data/data/AAPL.csv', parse_dates=['Date'])
df.sort_values('Date', inplace=True)
df.reset_index(drop=True, inplace=True)
```

Doing for AAPL tinker


```python
#Apply Technical Indicators with TA-Lib

df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)
df['EMA_20'] = talib.EMA(df['Close'], timeperiod=20)
df['RSI_14'] = talib.RSI(df['Close'], timeperiod=14)

macd, signal, hist = talib.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
df['MACD'] = macd
df['MACD_Signal'] = signal
df['MACD_Hist'] = hist
#Financial Metrics with PyNance
df['Daily_Return'] = df['Close'].pct_change()
df['Volatility_20'] = df['Daily_Return'].rolling(window=20).std()
df['Cumulative_Return'] = (1 + df['Daily_Return']).cumprod()

```

```python
#visualization of Closing Price with Moving Averages
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['SMA_20'], label='SMA 20')
plt.plot(df['Date'], df['EMA_20'], label='EMA 20')
plt.title("Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
# RSI
plt.figure(figsize=(14,4))
plt.plot(df['Date'], df['RSI_14'], label='RSI 14', color='purple')
plt.axhline(70, color='red', linestyle='--')   # Overbought
plt.axhline(30, color='green', linestyle='--') # Oversold

plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.show()

#Cumulative Return
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Cumulative_Return'], color='blue')
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.show()

#MACD

plt.figure(figsize=(14,4))
plt.plot(df['Date'], df['MACD'], label='MACD', color='blue')
plt.plot(df['Date'], df['MACD_Signal'], label='Signal', color='red')
plt.bar(df['Date'], df['MACD_Hist'], label='MACD Hist', color='gray')
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()

# Daily returns
plt.figure(figsize=(14,4))
plt.plot(df['Date'], df['Daily_Return'], color='purple')
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
    
# Rolling volatility
plt.figure(figsize=(14,4))
plt.plot(df['Date'], df['Volatility_20'], color='red')
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.show()
```
